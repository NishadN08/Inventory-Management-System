```python
#Import the json library
import json
```

```python
#Open the record.json file
fd = open("record.json",'r')
rec = fd.read()
fd.close()

#Loading the json file to perform Operations in python
record = json.loads(rec)
```

Main code starts from here...

```python
##Asking values to the customer
print("---------------Welcome to Shri Stores-------------\n")
print("******'Press 1' If you want to Add a product******")
print("****'Press 2' If you want to delete a product*****")
print("**********'Press 0' If you want to exit***********")
n = int(input())
#If user wants to Add a product
if n == 1:
    prod_id = input("Enter the product id:")
    name = input("Enter the name of product:")
    price = int(input("Enter the price:"))
    man_date=input("Enter the Manufactoring date of the product:")
    exp_date=input("Enter the Expiry date of the product:")
    quan = int(input("Enter the quantity:"))
    prod =[]
    for i in record:
        p = i.split(',') #Splits the keys in the dict by commas
        prod.append(int(p[0])) #Make a list out of the keys
    #Checking if the prod-id is present in database
    ecount = prod.count(int(prod_id))
    if ecount > 0:
        #Updating the data
        record[prod_id] = {'name':name,'price':price,"man_date":man_date,"exp_date":exp_date,'quantity':record[prod_id]['quantity']+quan}
    else:
        #Inserting the data into our record
        record[prod_id] = {'name': name,'price': price,"man_date": man_date,"exp_date": exp_date,'quantity': quan}
    print("\n****The product",prod_id,"has been recorded****")
    print("\n*******Thank you Please visit again********")
    
    js = json.dumps(record) #dumps the dict in .json format

    #Updating our record.json file with the added product
    fd = open("record.json",'w')
    fd.write(js) #Used for writting in the file
    fd.close()
#If user wants to exit
elif n == 0:
    print("\n*******Thank you Please visit again********")
#If user wants to delete a product
elif n==2:
    d = input("Enter the prod_id of the product you want to delete:\t")
#     prod =[]
#     for i in record:
#         p = i.split(',') #Splits the keys in the dict by commas
#         prod.append(int(p[0])) #Make a list out of the keys
    #Checking if the prod-id is present in database
    ecount = prod.count(int(d))
    if ecount > 0:
        del record[d] #deleting the record
        print("\nThe product",d,"has been deleted from our record")
    else:
        print("\n###The product is not present in our record###")
#If user has entered Invalid Choice
else:
    print("\n###Invalid Choice###")
```
